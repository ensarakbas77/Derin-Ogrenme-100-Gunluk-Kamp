## 📚 Neler Öğrendik?

### Dekoratörler

| Dekoratör | Açıklama | Kullanım Amacı |
|---|---|---|
| **Temel Dekoratörler** | Diğer fonksiyonları değiştiren fonksiyonlar | Orijinal kodu değiştirmeden yeni özellik eklemek |
| `@property / @setter` | Metotlara attribute gibi erişim sağlar | Veri doğrulama, hesaplanmış özellikler |
| `@staticmethod` | Instance veya class erişimi gerektirmeyen metotlar | Bir class ile ilişkili yardımcı fonksiyonlar |
| `@classmethod` | İlk argüman olarak class'ı alan metotlar | Alternatif constructor'lar, class değişkenlerine erişim |
| `@abstractmethod` | Alt sınıflar tarafından implement edilmesi zorunlu metotlar | Arayüz tanımlama, implementasyonu garanti altına alma |
| `@overload` | Birden fazla imzaya sahip fonksiyonlar için tip ipuçları | Daha iyi tip kontrolü ve IDE desteği |
| `@final` | Metotların override edilmesini veya miras alınmasını engeller | Kritik metotların değiştirilmemesini sağlamak |
| `@override` | Override edilen metotları açıkça işaretler | Metot ezme hatalarını önceden yakalamak |

---

### 💡 En İyi Pratikler

- Veri doğrulama ve hesaplanmış değerler için **`@property`** kullan
- Yardımcı fonksiyonlar için **`@staticmethod`** kullan
- Alternatif constructor'lar için **`@classmethod`** kullan
- Net arayüzler tanımlamak için **`@abstractmethod`** kullan
- Niyetini açıkça belirtmek için **`@override`** kullan
- Dekoratörleri gereğinden fazla kullanma — kodun okunabilirliğini koru
- Dekoratörlerin amacını açıklamak için **docstring** eklemeyi ihmal etme

---

### 🎯 Ne Zaman Hangisini Kullanmalısın?
```
Doğrulama mı gerekiyor?          → @property + @setter
Yardımcı fonksiyon mu?           → @staticmethod
Fabrika metodu mu?               → @classmethod
Arayüz / sözleşme mi?           → @abstractmethod
Tip ipucu mu?                    → @overload
Override'ı engellemek mi?        → @final
Üst sınıf metodunu mu eziyorsun? → @override
```
